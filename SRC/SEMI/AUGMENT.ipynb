```python
import os
from PIL import Image
# xám mix color
def chuyen_anh_xam_3_kenh(duong_dan_anh_goc, duong_dan_anh_xam):
    """Chuyển đổi ảnh màu sang ảnh xám 3 kênh và lưu."""
    try:
        anh_goc = Image.open(duong_dan_anh_goc).convert('RGB')
        anh_xam = anh_goc.convert('L')
        anh_xam_3_kenh = Image.merge('RGB', (anh_xam, anh_xam, anh_xam))
        anh_xam_3_kenh.save(duong_dan_anh_xam)
        return True
    except Exception as e:
        print(f"Lỗi khi xử lý ảnh {duong_dan_anh_goc}: {e}")
        return False

def sao_chep_va_tao_anh_xam(thu_muc_goc, thu_muc_moi):
    """Duyệt thư mục, sao chép ảnh và tạo ảnh xám trong thư mục mới."""
    if not os.path.exists(thu_muc_moi):
        os.makedirs(thu_muc_moi)

    for ten_thu_muc_con in os.listdir(thu_muc_goc):
        duong_dan_thu_muc_con_goc = os.path.join(thu_muc_goc, ten_thu_muc_con)
        if os.path.isdir(duong_dan_thu_muc_con_goc):
            duong_dan_thu_muc_con_moi = os.path.join(thu_muc_moi, ten_thu_muc_con)
            os.makedirs(duong_dan_thu_muc_con_moi, exist_ok=True)
            for ten_file in os.listdir(duong_dan_thu_muc_con_goc):
                duong_dan_file_goc = os.path.join(duong_dan_thu_muc_con_goc, ten_file)
                if os.path.isfile(duong_dan_file_goc) and any(ten_file.lower().endswith(ext) for ext in ['.png', '.jpg', '.jpeg']):
                    duong_dan_file_moi = os.path.join(duong_dan_thu_muc_con_moi, ten_file)
                    # Sao chép ảnh gốc
                    try:
                        Image.open(duong_dan_file_goc).save(duong_dan_file_moi)
                        ten_file_khong_duoi = os.path.splitext(ten_file)[0]
                        duong_dan_anh_xam_moi = os.path.join(duong_dan_thu_muc_con_moi, f"{ten_file_khong_duoi}_xam.png")
                        # Tạo và lưu ảnh xám
                        chuyen_anh_xam_3_kenh(duong_dan_file_goc, duong_dan_anh_xam_moi)
                        print(f"Đã xử lý: {duong_dan_file_goc}")
                    except Exception as e:
                        print(f"Lỗi khi sao chép hoặc xử lý file {duong_dan_file_goc}: {e}")

# Thiết lập đường dẫn thư mục gốc và thư mục mới
thu_muc_goc = '/kaggle/input/ck-final/rain_ck_ful'  # Đường dẫn đến thư mục 'data' của bạn
thu_muc_moi = 'data_xam+COLOR+CK'  # Tên thư mục mới sẽ được tạo

# Gọi hàm để thực hiện công việc
sao_chep_va_tao_anh_xam(thu_muc_goc, thu_muc_moi)

print("Hoàn tất quá trình xử lý ảnh.")
```
